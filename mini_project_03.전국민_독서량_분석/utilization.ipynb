#### 오디오북 독서율

```python
audiobook = pd.read_excel('./data/오디오북_독서율__학생__20220721104901.xlsx')
audiobook
```

```python
audiobook = audiobook.drop([audiobook.columns[0]], axis=1)
audiobook = audiobook.iloc[:5]
audiobook = audiobook.drop([1])
audiobook = audiobook.reset_index(drop=True)
```

```python
#### 3개의 표 합치기
```

```python
m1 = pd.merge(paper, ebook, how='outer', on='통계분류(2)')
m1 = pd.merge(m1, audiobook, how='outer', on='통계분류(2)')
```

```python
m1
```

```python
m1 = m1.rename(columns={'2021_x':'종이'})
m1 = m1.rename(columns={'2021_y':'전자'})
m1 = m1.rename(columns={'2021':'오디오'})
m1
```

```python
m1 = m1.drop(0, axis=0)
m1 = m1.reset_index(drop=True)
m1
```

```python
m2 = m1
m2['사례'] = [1117, 1095, 1108]
m2
```

```python
m2['종이_1'] = (m2['사례'] * m2['종이']) * 0.01
m2['전자_1'] = (m2['사례'] * m2['전자']) * 0.01
m2['오디오_1'] = (m2['사례'] * m2['오디오']) * 0.01
m2
```

```python
m2 = m2.drop(m2.columns[[1,2,3,4]], axis=1)
m2
```

```python
#### 그래프
import matplotlib.pyplot as plt
import numpy as np
```

```python
year = ['초등학교', '중학교', '고등학교']
```

```python
# 한글 폰트 지정
plt.rc('font', family='Malgun Gothic')
```

```python
fig, ax = plt.subplots(figsize=(12,6))
bar_width = 0.25

index = np.arange(3)

# 각 연도별로 3개 샵의 bar를 순서대로 나타내는 과정, 각 그래프는 0.25의 간격을 두고 그려짐
b1 = plt.bar(index, m1['종이'], bar_width, alpha=0.4, color='red', label='종이')
b2 = plt.bar(index + bar_width, m1['전자'], bar_width, alpha=0.4, color='blue', label='전자')
b3 = plt.bar(index + 2 * bar_width, m1['오디오'], bar_width, alpha=0.4, color='green', label='오디오')

# x축 위치를 정 가운데로 조정하고 x축의 텍스트를 year 정보와 매칭
plt.xticks(np.arange(bar_width, 3 + bar_width, 1), year)
plt.title('청소년 종이/전자/오디오별 독서률 비교')

# x축, y축 이름 및 범례 설정
# plt.xlabel('', size = 13)
plt.ylabel('독서률', size = 13)
plt.legend()
plt.show()
```
