## 공공도서관

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

### 공공도서관데이터 불러오기

```python
library=pd.read_excel("./2022년('21년 실적) 공공도서관 통계데이터 컬럼수정.xlsx")
```

```python
library
```

### 어린이실 유무 데이터정제

```python
library['어린이실 유무']=library['어린이실 유무'].replace('유',1)
library
```

```python
library['어린이실 유무']=library['어린이실 유무'].replace('무',0)
library
```

### none값 정제

```python
library=library.replace('공공(일반)에만 있는 통계항목',1)
library
```

```python
library_sum=library.groupby(['지역','어린이실 유무']).sum()
library_sum
```

## 필요 컬럼 정제

```python
library_sum=library_sum.drop(['개관년도'],axis=1)
```

```python
library_sum.columns=['어린이','청소년','성인','방문자수','어린이서비스이용자','어린이자료수']
```

```python
library_sum
```

```python
library_child=library_sum[['어린이']]
library_child
```

## 지역별 어린이등록회원수와 어린이실 유무의 관계검정

```python
# 검증하기
result = stats.chisquare(library_child['어린이'])

print('한 집단 평균 검정(단일 표본검정)=', result)
print('카이제곱-검정 통계량 = %.3f, p-value=%.3f'%(result))
#(해석-검증결과)
# 1.p-value=0.000<0.05 귀무기설 기각
# 대립가설 채택.: 지역별 어린이시설이 있는곳이 어린이등록회원에 영향을 준다.
```

## 설비시스템과 방문자수 검증

```python
library_equip=library.groupby(['설비_시스템구축현황']).sum()
library_equip
```

```python
result = stats.chisquare(library_equip['이용자수_도서관방문자수'])

print('한 집단 평균 검정(단일 표본검정)=', result)
print('카이제곱-검정 통계량 = %.3f, p-value=%.3f'%(result))
#(해석-검증결과)
# 1.p-value=0.000<0.05 귀무기설 기각
# 대립가설 채택.: 설비시스템에 대한 선호도 차이가 있다.
```

# 개관년도 별 방문자수

```python
library_year=library.groupby(['지역','개관년도']).sum()
library_year
```

```python
result = stats.chisquare(library_year['어린이서비스이용자수'])

print('한 집단 평균 검정(단일 표본검정)=', result)
print('카이제곱-검정 통계량 = %.3f, p-value=%.3f'%(result))
#(해석-검증결과)
# 1.p-value=0.000<0.05 귀무기설 기각
# 대립가설 채택.: 지역별 개관년도에 따라  어린이서비스이용수는 차이가 있다.
```

## 운영방식에따른 도서관 방문자수

```python
library_manage=pd.read_excel("./2022년('21년 실적) 공공도서관 통계데이터 운영방식.xlsx")
```

```python
library_manage_col=library_manage[['지역','운영방식','이용자수_도서관방문자수']]
```

```python
library_manage_sum=library_manage_col.groupby(['지역','운영방식']).sum()
library_manage_sum
```

```python
result = stats.chisquare(library_manage_sum['이용자수_도서관방문자수'])

print('한 집단 평균 검정(단일 표본검정)=', result)
print('t-검정 통계량 = %.3f, p-value=%.3f'%(result))
#(해석-검증결과)
# 1.p-value=0.000<0.05 귀무기설 기각
# 대립가설 채택.: 운영방식에따라 방문자수가 차이가있다.
```

```python
library_count=library.groupby(['지역']).count()
library_count
```

```python
library_sort=library[['지역','회원등록자수_연령별회원등록_어린이','회원등록자수_연령별회원등록_청소년',
        '회원등록자수_연령별회원등록_성인']]
library_sort_sum=library_sort.groupby('지역').sum()
library_sort_sum=library_sort_sum.reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 그림 사이즈, 바 굵기 조정
fig, ax = plt.subplots(figsize=(15,8))
bar_width = 0.25 ## Figure 생성 
fig.set_facecolor('white') ## Figure 배경색 지정
plt.rc('font',family ='Malgun Gothic',size=15)
#데이터넣기
plt.title('2021년 공공도서관 지역별 등록 회원수')
plt.ylabel('등록 회원수')
index = np.arange(17)
# 각 연도별로 3개 샵의 bar를 순서대로 나타내는 과정, 각 그래프는 0.25의 간격을 두고 그려짐
plt.bar(index, library_sort_sum['회원등록자수_연령별회원등록_어린이'], bar_width, alpha=0.4, 
        color='red', label='어린이')
plt.bar(index + bar_width, library_sort_sum['회원등록자수_연령별회원등록_청소년'], bar_width, alpha=0.4, 
        color='blue', label='청소년')
plt.bar(index + bar_width + bar_width, library_sort_sum['회원등록자수_연령별회원등록_성인'], 
        bar_width, alpha=0.4, color='green', label='|성인')
# x축 위치를 정 가운데로 조정하고 x축의 텍스트
plt.xticks(np.arange(bar_width, 4 + bar_width, 1))
plt.xticks([i for i in range(0,17,1)],['강원','경기','경남','경북','광주','대구','대전','부산','서울','세종',
                      '울산','인천','전남','전북','제주','충남','충북']
,rotation=90)


plt.ticklabel_format(axis='y',style='plain')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```
